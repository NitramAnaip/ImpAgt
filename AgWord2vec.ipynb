```python
import gensim
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
model = gensim.models.Word2Vec.load('Word2Vec/AgWordVectors-300.model')
```

### Mots les plus similaires à un mot donné

```python
model.wv.most_similar(positive = 'fish', topn = 5)
```

### Un peu de sémantique

```python
vec = model.wv['France'] - model.wv['Paris'] + model.wv['Berlin']
model.wv.most_similar(positive = [vec], topn = 1)
```

### Représentation graphique des similarités à l'aide d'une ACP

```python
vocab_to_plot = ['agriculture', 'cnn', 'rnn', 'svm', 'iot', 'crop', 'farm', 'tomato', 'potato', 'wheat', 'sugar']
```

```python
vocab_vectors = model.wv[vocab_to_plot]
```

```python
pca = PCA(n_components = 2)
res = pca.fit_transform(vocab_vectors)
```

```python
colors = np.random.rand(len(vocab_to_plot))
sizes = 1000*np.random.rand(len(vocab_to_plot))

plt.scatter(res[:,0], res[:,1], c = colors, s = sizes, alpha = 0.3, cmap = 'viridis')
plt.colorbar()

for i, word in enumerate(vocab_to_plot):
	plt.annotate(word, xy=(res[i, 0], res[i, 1]))
plt.show()
```
